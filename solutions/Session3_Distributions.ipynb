```python
import random
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# uniform
a = np.random.randint(1, 7, size=10000, dtype=np.int32)

plt.figure(figsize=(5,3))
sns.histplot(a, bins=6)
```

```python
a = np.random.randint(1, 7, size=(2, 100_000), dtype=np.int32)
a = a.sum(axis=0)

plt.figure(figsize=(5,3))
sns.histplot(a, bins=11)
```

```python
a = np.random.randint(1, 7, size=(100, 100_000), dtype=np.int32)
a = a.sum(axis=0)

plt.figure(figsize=(5,3))
sns.histplot(a, bins=50, kde=True)
```

```python
# geometric distribution
a = np.random.geometric(p=1/3, size=10_000)

plt.figure(figsize=(5,3))
sns.histplot(a, bins=20)
```

```python
a = np.random.geometric(p=1/3, size=(100, 10_000))
a = a.sum(axis=0)

plt.figure(figsize=(5,3))
sns.histplot(a, bins=50)
```
