```python
import seaborn as sns
```

```python
df = sns.load_dataset("penguins").dropna()
df.shape
```

```python
df["male"] = (df["sex"] == "Male").astype(int)
```

```python
df.head(3)
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(7, 5))  # inches
sns.scatterplot(data=df, x="body_mass_g", y="flipper_length_mm",
                hue="species")  # , style="male")
plt.xlabel("body mass [g]")
plt.ylabel("flipper length [mm]")
plt.savefig("scatter_penguins.png", dpi=100) # dots per inch
```

```python
sns.pairplot(df, hue="species")
```

```python
sns.lmplot(data=df, x="body_mass_g", y="flipper_length_mm", hue="species")
```

```python
num = df[["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g", "male"]]

num.corr().round(2)
```

```python
c = num.corr()
sns.heatmap(c, annot=True, vmin=-1.0, vmax=1.0)
```
