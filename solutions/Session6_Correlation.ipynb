```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = sns.load_dataset("penguins").dropna()
df.shape
```

```python
df.head(3)
```

```python
plt.figure(figsize=(5, 5)) # inches
sns.scatterplot(data=df, x="body_mass_g", y="flipper_length_mm", hue="sex")
plt.xlabel("body mass [g]")
plt.ylabel("flipper length [mm]")
# plt.xscale("log")
plt.savefig("pingus.png", dpi=200)  # dots per inch
```

```python
sns.lmplot(data=df, x="body_mass_g", y="flipper_length_mm")
```

```python
numeric = df[["bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"]]
numeric.corr().round(2)
```

```python
c = numeric.corr()
sns.heatmap(c, annot=True, vmin=-1.0, vmax=1.0)
```
